```python
# I want to calculate the amount of conversion needed in terms of the block
# reward multiple needed neuter the qi percentage, if we assume 5 zone blocks
# every prime blocks, then and the target is 5% for qi percentage, i want to
# graph the number of multiples of the block reward needed to neuter the qi
# percentage controller
import matplotlib.pyplot as plt
import numpy as np

# Parameters
zone_blocks_per_prime = 5  # 5 zone blocks per prime block
target_qi_percentage = 10
block_reward = 10

# Range of multiples to test
multiples = list(range(0, 101))

total_reward = block_reward * zone_blocks_per_prime 

block_reward_multiple_needed_quai_to_qi = []
block_reward_multiple_needed_qi_to_quai = []

for m in multiples:
    if m < target_qi_percentage:
        quai_to_qi = (target_qi_percentage - m) / 100 * total_reward
        qi_to_quai = 0
    elif m > target_qi_percentage:
        quai_to_qi = 0
        qi_to_quai = (m - target_qi_percentage) / 100 * total_reward
    else:  # m == target_qi_percentage
        quai_to_qi = 0
        qi_to_quai = 0
    
    block_reward_multiple_needed_quai_to_qi.append(quai_to_qi)
    block_reward_multiple_needed_qi_to_quai.append(qi_to_quai)

plt.figure(figsize=(10, 6))
plt.plot(multiples, block_reward_multiple_needed_quai_to_qi, label="Block reward multiple needed (Quai to Qi)", linewidth=2)
plt.plot(multiples, block_reward_multiple_needed_qi_to_quai, label="Block reward multiple needed (Qi to Quai)", linewidth=2)
plt.axhline(0, color='gray', linestyle='--', alpha=0.7)
plt.axvline(target_qi_percentage, color='red', linestyle=':', label=f'Target Qi % ({target_qi_percentage}%)')
plt.xlabel("Percentage of Qi taken")
plt.ylabel("Multiple of block reward needed to neuter")
plt.title("Block Reward Multiples Needed for 4000 prime blocks to Neuter Qi Percentage")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
